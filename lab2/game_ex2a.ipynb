```python
import pandas as pd
import matplotlib.pyplot  as plt
import numpy as np
import random
import time
from IPython.display import clear_output


    

exit_game = False
while not exit_game:
    
    number_of_points_1 = random.randint(40,50)
    number_of_points_2 = random.randint(40,50)
    winner = '0'
    if number_of_points_1 < number_of_points_2:
        winner = '2'
    else:
        winner = '1'
  
    #create two dataframes with random points
    df1 = pd.DataFrame(np.random.random_sample(size=(number_of_points_1, 2)), columns=list('xy'))
    df2 = pd.DataFrame(np.random.random_sample(size=(number_of_points_2, 2)), columns=list('xy'))
    plt.scatter(df1.x, df1.y)
    plt.show()
    
    plt.scatter(df2.x, df2.y, color = 'green')
    plt.show()
    print("Which plot has more points? Take a guess. Press 1 if you think the top one, 2 for the bottom.")
    print("If you think they are equal press 0")
    print("If you want to exit type x")
    
    command = input("Enter the command and press enter")
    if command == 'x':
        break
        
    if command == winner:
        print("CORRECT!")
    else:
        print("INCORRECT!")
    print("Top plot:" ,number_of_points_1, "Bottom plot: ", number_of_points_2)
   
    print("..... waiting 10 second for the next game......")
    time.sleep(10)
    plt.clf()
    plt.cla()
    plt.close()

    clear_output(wait=True)

 
    
```
