```python
import pandas as pd
import matplotlib.pyplot  as plt
import numpy as np

df1 = pd.DataFrame(np.random.random_sample(size=(14, 1)), columns=list('y'))
df1['x'] = np.arange(len(df1))
df1['x'] = range(len(df1))
df2 = pd.DataFrame(np.random.random_sample(size=(14, 1)), columns=list('y'))
df2['x'] = np.arange(len(df2))
df2['x'] = range(len(df2))

plt.scatter(df1.x,df1.y, marker='o')
plt.scatter(df2.x, df2.y, marker ='s')
plt.show()

plt.plot(df1.x,df1.y, marker='o')
plt.plot(df2.x, df2.y, marker ='s')
plt.show()

```
