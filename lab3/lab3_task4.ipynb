### b) The Car dataset is a classic. Use the dataset auto-mpg.csv [1] by loading it into a data frame. There are nine attributes. What is an interesting aspect? You could use visualizations exploratively here.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r'C:\Users\User\OneDrive\Uni\Semester 3 - ERASMUS\Data Visualisation\Labs\Labs 3\Exercise material-20241125\auto-mpg.csv'  
auto_mpg = pd.read_csv(file_path)

print(auto_mpg)


auto_mpg['horsepower'] = pd.to_numeric(auto_mpg['horsepower'], errors='coerce')
auto_mpg_clean = auto_mpg.dropna()

# Scatter plot: mpg vs. weight
plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
sns.scatterplot(x='weight', y='mpg', data=auto_mpg_clean)
plt.title('MPG against Weight')
plt.xlabel('Weight')
plt.ylabel('MPG')

# Scatter plot: mpg vs. horsepower
plt.subplot(2, 2, 2)
sns.scatterplot(x='horsepower', y='mpg', data=auto_mpg_clean)
plt.title('MPG against Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('MPG')

# Line plot: mpg trend over model years
plt.subplot(2, 2, 3)
sns.lineplot(x='model year', y='mpg', data=auto_mpg_clean, errorbar=None)
plt.title('MPG Trend against Model Years')
plt.xlabel('Model Year')
plt.ylabel('MPG')

# Box plot: mpg by origin
plt.subplot(2, 2, 4)
sns.boxplot(x='origin', y='mpg', data=auto_mpg_clean)
plt.title('MPG by Origin')
plt.xlabel('Origin (1=USA, 2=Europe, 3=Japan)')
plt.ylabel('MPG')

plt.figure(figsize=(12, 6))

# Plot each variable on the same graph
sns.lineplot(x='model year', y='horsepower', data=auto_mpg_clean, label='Horsepower', color='blue', errorbar=None)
sns.lineplot(x='model year', y='acceleration', data=auto_mpg_clean, label='Acceleration', color='red', errorbar=None)

# Add title and labels
plt.title('Horsepower and Acceleration against Model Years')
plt.xlabel('Model Year')
plt.ylabel('Value')

# Add legend to distinguish lines
plt.legend(title='Attributes')

plt.figure(figsize=(12, 6))

# Plot both displacement and horsepower on the same graph
sns.lineplot(x='model year', y='displacement', data=auto_mpg_clean, label='Displacement', color='blue', errorbar=None)
sns.lineplot(x='model year', y='horsepower', data=auto_mpg_clean, label='Horsepower', color='orange', errorbar=None)

# Add title and labels
plt.title('Displacement and Horsepower against Model Years')
plt.xlabel('Model Year')
plt.ylabel('Value')

# Add legend to distinguish the lines
plt.legend(title='Attributes')

# Display the plot
plt.tight_layout()
plt.show()
```

1. MPG against Weight: There is a clear negative correlation; heavier cars tend to have lower fuel efficiency.
2. MPG against Horsepower: The same occurs here, cars with higher horsepower tend to have lower MPG, showing that more powerful engines consume more fuel.
3. MPG Trend against Model Years: There is a noticeable improvement in fuel efficiency over the years, which imply technological advancements.
4. MPG by Origin: Cars from Japan seem to always have a better MPG, while those from the USA tend to have lower MPG.
5. Horsepower and Acceleration against Model Years: As time went by, the acceleration of the cars is seemed to stay relatively stable, while the horsepower decreased immensely, also indicating technological advancements.
6. Displacement and Horsepower against Model Years: Another visualization which reveals how technology has improved, due to smaller engine sizes only decreased the horsepower by a small margin.  

### Visualize one aspect three times with varying encodings. If you use color, do so deliberately and explain your choices. Discuss if your perception agrees on the effectiveness of the channels and which visualization is best.  

```python
# visualization 1: Scatterplot
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.scatterplot(x='weight', y='mpg', data=auto_mpg_clean)
plt.title('MPG vs Weight (Position Encoding)')
plt.xlabel('Weight')
plt.ylabel('MPG')

# visualization 2: Color Gradient
plt.subplot(1, 3, 2)
sns.scatterplot(x='weight', y='displacement', hue='mpg', palette='coolwarm', data=auto_mpg_clean)
plt.title('MPG via Color Gradient')
plt.xlabel('Weight')
plt.ylabel('Displacement')
plt.legend(title='MPG')

# visualization 3: Bubble Plot
plt.subplot(1, 3, 3)
sns.scatterplot(x='weight', y='displacement', size='mpg', sizes=(10, 200), data=auto_mpg_clean, alpha=0.6)
plt.title('MPG via Bubble Size')
plt.xlabel('Weight')
plt.ylabel('Displacement')
plt.legend(title='MPG', loc='upper left')

plt.tight_layout()
plt.show()
```

1. MPG vs Weight (Position Encoding): makes the trend between MPG and Weight easily comparable.
2. MPG via Colour Gradient: The warmer the colour indicates the greater the MPG. This is done to make the plot more visually appealing, and to help understand the graph.
3. MPG via Bubble Size: This uses the size of the bubbles to show a higher MPG.

The best visualization is probably the scatterplot, as it is the most useful plot when trying to get precise comparisons.
