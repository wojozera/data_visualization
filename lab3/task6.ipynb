```python
import matplotlib.pyplot as plt
import numpy as np

# Data
countries = ['USA', 'Australia']
group1_values = [3.8, 1.3]  
group2_values = [7.2, 0.6]  

# Define bar positions
x = np.arange(len(countries)) 
bar_width = 0.5  

# Create the plot
fig, ax = plt.subplots(figsize=(6, 4))

# Plot bars for the two groups
bar1 = ax.bar(x - bar_width / 2, group1_values, bar_width, label='Group 1', color='blue')
bar2 = ax.bar(x + bar_width / 2, group2_values, bar_width, label='Group 2', color='red')


for bars in bar1:
    ax.text(bars.get_x() + bars.get_width() / 2, bars.get_height() + 0.1, f"${bars.get_height():.1f}K", ha='center')

for bars in bar2:
    ax.text(bars.get_x() + bars.get_width() / 2, bars.get_height() + 0.1, f"${bars.get_height():.1f}K", ha='center')

ax.set_xticks(x)
ax.set_xticklabels(countries)
ax.set_ylabel('Values')

# Show the plot
plt.show()
```
