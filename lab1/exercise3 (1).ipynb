```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Load data

```python
df = pd.read_csv("kochin2.CSV",decimal=',',delimiter=';')
df.set_index('YEAR')
```

### Plot the employment

```python
df.plot(x='YEAR', y= 'UNEMP')
```

Plot unemployment / benefits

```python

fig, axs = plt.subplots()
axs.plot(df.YEAR,df.UNEMP)
axs.plot(df.YEAR,df.BWRATIO * 50)
axs.set_title('UNEMP/ BWRATIO')
axs.legend(['UNEMP', 'BWRATIO'])
```

Unemployment / Demand

```python
fig, axs = plt.subplots()
axs.plot(df.YEAR,df.UNEMP)
axs.plot(df.YEAR,df.DEMAND)
axs.set_title('UNEMP/ DEMAND')
axs.legend(['UNEMP', 'DEMAND'])
```

### Plot both at the same time

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(df.YEAR,df.UNEMP, 'g-')
ax2.plot(df.YEAR,df.DEMAND, 'b-')

ax1.set_xlabel('YEAR')
ax1.set_ylabel('UNEMP', color='g')
ax2.set_ylabel('DEMAND', color='b')

plt.show()
```
