```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt


datasaurus_dozen = pd.read_csv('DatasaurusDozen.tsv', sep='\t')
#aggregations for each gruop and calculating the statistics
results  = datasaurus_dozen.groupby('dataset').agg({'x':['mean', 'std', 'var'], 'y':['mean', 'std', 'var'] })
print(results)

#getting all the unique columns
columns = datasaurus_dozen['dataset'].unique()
print(columns)

#scatter plot for each group
for column in columns:
    one_of_dozen = datasaurus_dozen.loc[datasaurus_dozen['dataset'] == column]
    plt.scatter(one_of_dozen.x, one_of_dozen.y)
    plt.show()
    


```
